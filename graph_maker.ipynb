```python
import matplotlib.pyplot as plt
import subprocess
import os
```

```python
script_dir = os.path.dirname(__file__)
rel_path = "output.txt"
file = os.path.join(script_dir, rel_path)
```

```python
import matplotlib.pyplot as plt

# Open input file and read in all lines
with open(file, 'r') as f:
    lines = f.readlines()

# Parse coordinates from first line
coords = lines[0].strip().split()
x_coords = [float(c) for c in coords[0::2]]
y_coords = [float(c) for c in coords[1::2]]
x_coords.append(x_coords[0])
y_coords.append(y_coords[0])

# Plot first polygon
#plt.subplot(2, 1, 1)
plt.plot(x_coords, y_coords)
plt.title('First Polygon')
plt.show()
```

```python
# Plot remaining polygons
for line in lines[1:]:
    coords = line.strip().split()
    x_coords = [float(c) for c in coords[0::2]]
    y_coords = [float(c) for c in coords[1::2]]
    x_coords.append(x_coords[0])
    y_coords.append(y_coords[0])
    #plt.subplot(2, 1, 2)
    plt.plot(x_coords, y_coords)
plt.title('Remaining Polygons')

# Show plots
plt.show()
```
